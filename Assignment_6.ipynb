```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('seaborn-v0_8')
from matplotlib import colormaps


```

```python
a = mr.MesaData(file_name='Assignment_6_LOGS/history.data')
l = mr.MesaLogDir('Assignment_6_LOGS')


# load the profile with PROFILE number 12
p_1 = l.profile_data(profile_number=1)

# load the last profile saved (largest model number)
p_2 = l.profile_data()
```

# Assignment-6

# qs.a

```python
plt.plot(p_1.r_div_R,p_1.logRho,label='near ZAMS')
plt.plot(p_2.r_div_R,p_2.logRho,label='near He burning phase')
plt.legend(loc='upper right')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('log_{10} (r/R)')
plt.ylabel('log_{10} (rho) ')
plt.show()
```
